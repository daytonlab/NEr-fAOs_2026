## Inspect matrixes AnnData

```python
import numpy as np
from scipy.sparse import issparse

def matrix_summary(adata):
    """
    Prints min, max, and mean for expression matrix, counts layer, raw matrix,
    and scaled layer of an AnnData object.

    Parameters:
        adata: AnnData
    """

    # Helper function to compute min, max, mean
    def summary(matrix, name):
        if issparse(matrix):
            matrix = matrix.toarray()
        flat = matrix.flatten()
        mean_value = flat.mean()
        print(f"{name}: min={flat.min()}, max={flat.max()}, mean={mean_value}")

    # Expression matrix
    summary(adata.X, "Expression matrix (X)")

    # Counts layer
    if 'counts' in adata.layers:
        summary(adata.layers['counts'], "Counts layer matrix (layers['counts'])")
    else:
        print("Counts layer not found in adata.layers.")

    # Raw matrix
    if adata.raw is not None:
        summary(adata.raw.X, "Raw matrix (raw.X)")
    else:
        print("Raw matrix not found in adata.raw.")

    # Scaled layer
    if 'scaled' in adata.layers:
        summary(adata.layers['scaled'], "Scaled layer matrix (layers['scaled'])")
    else:
        print("Scaled layer not found in adata.layers.")
```

## Building *AnnData* object

### Cell metadata

We will include Region and Time Points based on the *orig.ident* value for each cell

```python
# Load cell metadata
cell_meta = pd.read_csv('./../data/Conchola2023/Conchola2023FetalEpithelium_metadata.csv', index_col= 12)
cell_meta.rename(
    columns= {
        'orig.ident' : 'Patient ID'
    },
    inplace= True
)
cell_meta['Tissue'] = 'Fetal Epithelium'

# Map Region to Patient ID suffix
cell_meta['Region'] = (
    cell_meta['Patient ID']
    .str.extract(r'd(airway|distal|trach)$')[0]  # final letters of Patient ID
    .str.lower()
    .map({
        'distal' : 'Distal',
        'airway': 'Airway',
        'trach': 'Trachea',
    })
)

# Add time points
cell_meta['Time point'] = cell_meta['Patient ID'].replace(
    {
        'f103dairway' : 'week15',
        'f103ddistal' : 'week15',
        'f125dairway' : 'week18',
        'f125ddistal' : 'week18',
        'f132ddistal' : 'week19',
        'f59dairway' : 'week8',
        'f59ddistal' : 'week8',
        'f80dairway' : 'week11',
        'f80ddistal' : 'week11',
        'f103dtrach' : 'week15',
        'f145dtrach' : 'week21'
    }
)

#Save Sample ID information
cell_meta['Sample ID'] = cell_meta['Patient ID']


# Clean up Patient ID because we have stored the region and time point independently
cell_meta['Patient ID'] = cell_meta['Patient ID'].replace(
    {
        'f103dairway' : 'f103',
        'f103ddistal' : 'f103',
        'f125dairway' : 'f125',
        'f125ddistal' : 'f125',
        'f132ddistal' : 'f132',
        'f59dairway' : 'f59',
        'f59ddistal' : 'f59',
        'f80dairway' : 'f80',
        'f80ddistal' : 'f80',
        'f103dtrach' : 'f103',
        'f145dtrach' : 'f145'
    }
)

cell_meta
```

## Gene Annotation

***FROM THIS POINT FORWARD, WE WILL WORK WITH THE TRIMMED ANNDATA,  GENERATED IN PREVIOUS BLOCK***

We want to be able to merge this data with ours to compare them. 

First we need to make sure genes are annotated the same way in both datasets.

```python
# Load object
adata_fetal = sc.read_h5ad('./../data/Conchola2023/Conchola2023_FetalEpithelium.h5ad')
adata_fetal
```

***NOTICE*** we do not have ENS_IDs for the genes

```python
adata_fetal.var
```

This means we will need to map the gene names to their original ENS_ID (*gencode.v19*) and then map those to our names (*gencode.v44*)

```python
# Fxn to parse gencode annotation into dict
def gencode_parser(filepath):
    # Use regular expressions to find gene_id and gene_name regardless of number of columns in input annotation
    import re
    # gene_id_pattern = re.compile(r'gene_id\s+"([^"]+)')
    gene_id_pattern = re.compile(r'gene_id\s+"(ENS[A-Z]+[0-9]+)')
    gene_name_pattern = re.compile(r'gene_name\s+"([^"]+)"')

    with open(filepath, 'r') as f:
        gtf = {
            gene_id_pattern.search(line).group(1) : gene_name_pattern.search(line).group(1)
            for line in f
            if not line.startswith('#')
        }
    return gtf
```

### *Gencode v19* mapping

```python
# Parse v19 gene annotation and save and inverted dict where:
    # gene_name : ENS_ID
inv_gtf_v19 = {
    v : k
    for k, v in gencode_parser('./../data/annotations_aux_files/gencode.v19.annotation.gtf').items()
}

# Store v19 gene name
adata_fetal.var['gencode_og_name'] = list(adata_fetal.var_names)

# Map v19 gene names to their ENS_IDs
adata_fetal.var['gencode_og_id'] = adata_fetal.var_names.map(inv_gtf_v19)

del inv_gtf_v19
adata_fetal.var
```

***NOTICE*** there are some entries with NaNs in the gencode_og_id column.

This means these gene names, as they are, did not map to any ENS_ID

```python
adata_fetal.var[adata_fetal.var['gencode_og_id'].isna()]
```

### *Gencode v44* mapping

```python
# Parse v44 gene annotation
gtf_v44_annots = gencode_parser('./../data/annotations_aux_files/gencode.v44.annotation.gtf')

# We now map the v19 ENS_ID to the v44 ENS_ID's gene name
adata_fetal.var['gencode_v44_name'] = adata_fetal.var['gencode_og_id'].map(gtf_v44_annots)

del gtf_v44_annots
adata_fetal.var
```

***
NOTICE NaN values can be either for one of two reasons:
    - No v19 ENS_ID to map to v44 ENS_ID
    - A v19 ENS_ID is not included in the v44 annotation, therefore no mapping possible
***

```python
adata_fetal.var[adata_fetal.var['gencode_v44_name'].isna()]
```

```python
# We will assign the original gene name to those we were unable to map
for var_name in adata_fetal.var.index:
    if pd.isna(adata_fetal.var.loc[var_name, 'gencode_v44_name']):
        adata_fetal.var.loc[var_name, 'gencode_v44_name'] = adata_fetal.var.loc[var_name, 'gencode_og_name']

# Display NaN values for v44_names
adata_fetal.var[adata_fetal.var['gencode_v44_name'].isna()]
```

```python
# We can now make this v44 names the index for the data
adata_fetal.var_names = adata_fetal.var['gencode_v44_name']
```

### Handling duplicated gene entries

We now have to make sure there are no duplicates.

**Why could there be duplicates?** 

> Because one gene could have the same gene name with two different ENS_IDs in the original dataset.

```python
# Display duplicated entries
duplicate_var_names = adata_fetal.var_names[adata_fetal.var_names.duplicated()]
adata_fetal.var.loc[adata_fetal.var['gencode_v44_name'].isin(duplicate_var_names), :].sort_index()
```

**Generate**
- AnnData with only duplicated entries

```python
# Filter by duplicated var_names
adata_dup = adata_fetal[:, adata_fetal.var['gencode_v44_name'].isin(duplicate_var_names)].copy()
adata_dup
```

```python
# We will split the duplicated var_names object into doublets for each unique var_name
duplicates_dict = {}
duplicates_names = list(adata_dup.var_names)
for var_name in duplicates_names:
    duplicates_dict[var_name] = adata_dup[:, var_name]

del duplicates_names
del var_name
duplicates_dict
```

**NOTICE:** in order to have only one entry for each unique gene, we will sum both their counts value

We will recreate an AnnData object for each duplicated entry with the collpased data and merged metadata

```python
ann_data_dict_sum = {}

for anndata_key, anndata_value in duplicates_dict.items():

    var_name = anndata_value.var_names[:1]
    
    # Sum the expression values for genes for the current AnnData object 
    summed_expression = csr_matrix(anndata_value.X.sum(axis= 1))

    # Create a new AnnData object to store the summed expression and all merged metadata
    ann_data_dict_sum[anndata_key] = ad.AnnData(
        X = summed_expression,
        obs= anndata_value.obs.copy(),
        var= {
            'gencode_og_id' : ', '.join(list(map(str, set(anndata_value.var.loc[:, 'gencode_og_id'])))),
            'gencode_og_name' : ', '.join(list(map(str, set(anndata_value.var.loc[:, 'gencode_og_name'])))),
            'gencode_v44_name' : ', '.join(list(map(str, set(anndata_value.var.loc[:, 'gencode_v44_name'])))),
            'summed_genes' : True
        }
    )
    # Specify a new name for the collapsed gene axis
    ann_data_dict_sum[anndata_key].var_names = var_name

adata_summed_dups = ad.concat(
    list(ann_data_dict_sum.values()),
    axis= 1,
    join= 'outer',
    merge= 'unique'
)

# Remove all objects we no longer need
del duplicates_dict
del anndata_value
del anndata_key
del summed_expression
del var_name
del ann_data_dict_sum
del adata_dup

adata_summed_dups
```

We will now create a separate AnnData with only unique entries (to merge in following steps with our collapsed-entries AnnData)

```python
# Filter out duplicated var_names
adata_unique = adata_fetal[:, ~adata_fetal.var['gencode_v44_name'].isin(duplicate_var_names)].copy()
adata_unique.var['summed_genes'] = False

# These dimensionality reductions no longer represent the current data
del adata_unique.obsm['X_pca']
del adata_unique.obsm['X_umap']

# Plotting attributes not necessary
del adata_unique.uns['Region_colors']
del adata_unique.uns['annotation_colors']

adata_unique, bool(adata_unique.var_names.duplicated().any())
```

We concat both AnnData objects to have a final one without duplicates and with full metadata

```python
adata_final = ad.concat(
    [adata_unique, adata_summed_dups],
    axis= 1,
    join= 'outer',
    merge= 'unique'
)
del adata_unique
del adata_summed_dups

adata_final
```

***NOTICE:*** the size of our original AnnData object compared to our final AnnData

There is only a 17-genes difference (the duplicates)

```python
adata_fetal, adata_final
```

## Exploration of annotated scRNA-seq object

- Add useful metadata
- Preprocess the data:
    - Perform a quality check (observational only)
        - Mitochondrial and ribosomal genes
        - Counts-based quality
    - Normalization
    - Score and regress out cells based on gene cell cycle scores

```python
adata_fetal = sc.read_h5ad('./../data/Conchola2023/Conchola2023_FetalEpithelium_annotated.h5ad')
adata_fetal.obs.drop(
    ['nCount_RNA', 'nFeature_RNA', 'percent.mt', 'percent.ribo', 'S.Score', 'G2M.Score', 'Phase'],
    axis= 1,
    inplace= True
)
adata_fetal.obs['Technique'] = '10x'
adata_fetal.obs['Reference'] = '(Conchola et al., 2023)'

adata_fetal
```

## Normalization & Highly Variable Genes

```python
# Save raw data
adata_fetal.layers['counts'] = csr_matrix(adata_fetal.X)

# Normalize counts per cell.
# Normalize each cell by total counts over all genes, so that every cell has the same total count after normalization
sc.pp.normalize_total(adata_fetal, target_sum=1e4)

# Logarithmize the data matrix.
# Computes X = log (X=1), where X denotes the natural logarithm unless a different base is given.
sc.pp.log1p(adata_fetal)

adata_fetal.raw = adata_fetal.copy()
adata_fetal
```

```python
matrix_summary(adata_fetal)
```

```python
sc.pp.highly_variable_genes(adata_fetal, layer= None, n_top_genes= None, min_disp= 0.5, min_mean= 0.0125,
                                max_mean= 3, span= 0.3, n_bins= 20, flavor= 'seurat', 
                                subset= False, inplace= True, batch_key= None, check_values= True)
# batch_key set to None
# If specified, highly-variable genes are selected within each batch separately and merged. 
# This simple process avoids the selection of batch-specific genes and acts as a lightweight batch correction method. 

sc.pl.highly_variable_genes(adata_fetal) 
```

## Scaling of the data

```python
# Scale data to unit variance and zero mean store results in layer
# Variables (genes) that do not display any variation (are constant across all observations) 
# are retained and set to 0 during this operation. In the future, they might be set to NaNs.

adata_fetal.layers["scaled"] = sc.pp.scale(adata_fetal, copy=True).X.copy()
adata_fetal
```
