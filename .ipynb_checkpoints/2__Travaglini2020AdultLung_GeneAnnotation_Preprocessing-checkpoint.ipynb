## Inspect matrixes AnnData

```python
import numpy as np
from scipy.sparse import issparse

def matrix_summary(adata):
    """
    Prints min, max, and mean for expression matrix, counts layer, raw matrix,
    and scaled layer of an AnnData object.

    Parameters:
        adata: AnnData
    """

    # Helper function to compute min, max, mean
    def summary(matrix, name):
        if issparse(matrix):
            matrix = matrix.toarray()
        flat = matrix.flatten()
        mean_value = flat.mean()
        print(f"{name}: min={flat.min()}, max={flat.max()}, mean={mean_value}")

    # Expression matrix
    summary(adata.X, "Expression matrix (X)")

    # Counts layer
    if 'counts' in adata.layers:
        summary(adata.layers['counts'], "Counts layer matrix (layers['counts'])")
    else:
        print("Counts layer not found in adata.layers.")

    # Raw matrix
    if adata.raw is not None:
        summary(adata.raw.X, "Raw matrix (raw.X)")
    else:
        print("Raw matrix not found in adata.raw.")

    # Scaled layer
    if 'scaled' in adata.layers:
        summary(adata.layers['scaled'], "Scaled layer matrix (layers['scaled'])")
    else:
        print("Scaled layer not found in adata.layers.")
```

### Aux Fxns

```python
# Define functions to explore the size of the object and its components.
def sizeof_fmt(num, suffix="B"):
    for unit in ['','K','M','G','T']:
        if abs(num) < 1024.0:
            return f"{num:3.1f}{unit}{suffix}"
        num /= 1024.0
    return f"{num:.1f}P{suffix}"
def adata_size_breakdown(adata, is_raw: bool= False):
    total = 0
    print("\t=== Aprox. size of elements ===")
    # X
    if hasattr(adata, "X") and adata.X is not None:
        size = adata.X.data.nbytes if hasattr(adata.X, "data") else adata.X.nbytes
        total += size
        print(f"X: {sizeof_fmt(size)}")
    if not is_raw:
        # layers
        for key, layer in adata.layers.items():
            size = layer.data.nbytes if hasattr(layer, "data") else layer.nbytes
            total += size
            print(f"layers[{key}]: {sizeof_fmt(size)}")
        # obsm
        for key, val in adata.obsm.items():
            size = val.data.nbytes if hasattr(val, "data") else val.nbytes
            total += size
            print(f"obsm[{key}]: {sizeof_fmt(size)}")
        # varm
        for key, val in adata.varm.items():
            size = val.data.nbytes if hasattr(val, "data") else val.nbytes
            total += size
            print(f"varm[{key}]: {sizeof_fmt(size)}")
        # uns
        import pickle
        for key, val in adata.uns.items():
            try:
                size = len(pickle.dumps(val))
                total += size
                print(f"uns[{key}]: {sizeof_fmt(size)}")
            except Exception:
                pass
    print(f"\t=== Aprox. total size: {sizeof_fmt(total)} ===\n")
```

## Gene Annotation

***FROM THIS POINT FORWARD, WE WILL WORK WITH THE TRIMMED ANNDATA,  GENERATED IN PREVIOUS BLOCK***

We want to be able to merge this data with ours to compare them. 

First we need to make sure genes are annotated the same way in both datasets.

```python
adata_facs_trimmed = sc.read_h5ad('./data/Travaglini2020/Travaglini2020_FacsNormalLungBlood_trimmed.h5ad')
adata_10x_trimmed = sc.read_h5ad('./data/Travaglini2020/Travaglini2020_10xNormalLungBlood_trimmed.h5ad')

adata_facs_10x_trimmed = ad.concat(
    [adata_facs_trimmed, adata_10x_trimmed],
    axis= 0,
    join= 'outer'
)

# Delete individual anndatas
del adata_facs_trimmed
del adata_10x_trimmed

adata_facs_10x_trimmed, sc.pl.tsne(adata_facs_10x_trimmed, color=['Technique'], vmin= 0, vmax= 'p96')
```

```python
# We will seek to work with raw counts in order to process them ourselves
del adata_facs_10x_trimmed.obsm
del adata_facs_10x_trimmed.uns
adata_facs_10x_trimmed.layers['previous_X'] = adata_facs_10x_trimmed.X
adata_facs_10x_trimmed.X = adata_facs_10x_trimmed.raw.X
del adata_facs_10x_trimmed.raw
adata_facs_10x_trimmed
```

```python
# Fxn to parse gencode annotation into dict
def gencode_parser(filepath):
    # Use regular expressions to find gene_id and gene_name regardless of number of columns in input annotation
    import re
    # gene_id_pattern = re.compile(r'gene_id\s+"([^"]+)"')
    gene_id_pattern = re.compile(r'gene_id\s+"(ENS[A-Z]+[0-9]+)')
    gene_name_pattern = re.compile(r'gene_name\s+"([^"]+)"')

    with open(filepath, 'r') as f:
        gtf = {
            gene_id_pattern.search(line).group(1) : gene_name_pattern.search(line).group(1)
            for line in f
            if not line.startswith('#')
        }
    return gtf
```

### Gencode *v29* mapping

```python
# Parse v29 gene annotation and save and inverted dict where:
    # gene_name : ENS_ID
inv_gtf_v19 = {
    v : k
    for k, v in gencode_parser('./data/annotations_aux_files/gencode.v29.annotation.gtf').items()
}

# Store v29 gene name
adata_facs_10x_trimmed.var['gencode_og_name'] = list(adata_facs_10x_trimmed.var_names)

# Map v19 gene names to their ENS_IDs
adata_facs_10x_trimmed.var['gencode_og_id'] = adata_facs_10x_trimmed.var_names.map(inv_gtf_v19)

del inv_gtf_v19
adata_facs_10x_trimmed.var
```

***NOTICE*** there are some entries with NaNs in the gencode_og_id column.

This means these gene names, as they are, did not map to any ENS_ID

```python
adata_facs_10x_trimmed.var[adata_facs_10x_trimmed.var['gencode_og_id'].isna()]
```

### Gencode *v44* mapping

```python
# Parse v44 gene annotation
gtf_v44_annots = gencode_parser('./data/annotations_aux_files/gencode.v44.annotation.gtf')

# We now map the v19 ENS_ID to the v44 ENS_ID's gene name
adata_facs_10x_trimmed.var['gencode_v44_name'] = adata_facs_10x_trimmed.var['gencode_og_id'].map(gtf_v44_annots)

del gtf_v44_annots
adata_facs_10x_trimmed.var
```

```python
"""
NOTICE NaN values can be either for one of two reasons:
    - No v19 ENS_ID to map to v44 ENS_ID
    - A v19 ENS_ID is not included in the v44 annotation, therefore no mapping possible
"""
adata_facs_10x_trimmed.var[adata_facs_10x_trimmed.var['gencode_v44_name'].isna()]
```

```python
# We will assign the original gene name to those we were unable to map
for var_name in adata_facs_10x_trimmed.var.index:
    if pd.isna(adata_facs_10x_trimmed.var.loc[var_name, 'gencode_v44_name']):
        adata_facs_10x_trimmed.var.loc[var_name, 'gencode_v44_name'] = adata_facs_10x_trimmed.var.loc[var_name, 'gencode_og_name']

# Display NaN values for v44_names
adata_facs_10x_trimmed.var[adata_facs_10x_trimmed.var['gencode_v44_name'].isna()]
```

```python
# We can now make this v44 names the index for the data
adata_facs_10x_trimmed.var_names = adata_facs_10x_trimmed.var['gencode_v44_name']
```

### Handling duplicated gene entries

We now have to make sure there are no duplicates.

**Why could there be duplicates?** 

> Because one gene could have the same gene name with two different ENS_IDs in the original dataset.

```python
# Display duplicated entries
duplicate_var_names = adata_facs_10x_trimmed.var_names[adata_facs_10x_trimmed.var_names.duplicated()]
adata_facs_10x_trimmed.var.loc[adata_facs_10x_trimmed.var['gencode_v44_name'].isin(duplicate_var_names), :].sort_index()
```

**NOTICE:** Different Original ENS_IDs but same name

**Generate**
- AnnData with only duplicated entries

```python
# Filter by duplicated var_names
adata_dup = adata_facs_10x_trimmed[:, adata_facs_10x_trimmed.var['gencode_v44_name'].isin(duplicate_var_names)].copy()
adata_dup
```

```python
# We will split the duplicated var_names object into doublets for each unique var_name
duplicates_dict = {}
duplicates_names = list(adata_dup.var_names)
for var_name in duplicates_names:
    duplicates_dict[var_name] = adata_dup[:, var_name].copy()

del duplicates_names
del var_name
duplicates_dict
```

**NOTICE:** in order to have only one entry for each unique gene, we will sum both their counts value

We will recreate an AnnData object for each duplicated entry with the collpased data and merged metadata

```python
ann_data_dict_sum = {}

for anndata_key, anndata_value in duplicates_dict.items():

    var_name = anndata_value.var_names[:1]
    
    # Sum the expression values for genes for the current AnnData object 
    summed_expression = csr_matrix(anndata_value.X.sum(axis= 1))
                        
    # Create a new AnnData object to store the summed expression and all merged metadata
    ann_data_dict_sum[anndata_key] = ad.AnnData(
        X = summed_expression,
        obs= anndata_value.obs.copy(),
        layers= {
            'previous_X' : csr_matrix(anndata_value.layers['previous_X'].sum(axis= 1))
        },
        var= {
            'gencode_og_id' : ', '.join(list(map(str, set(anndata_value.var.loc[:, 'gencode_og_id'])))),
            'gencode_og_name' : ', '.join(list(map(str, set(anndata_value.var.loc[:, 'gencode_og_name'])))),
            'gencode_v44_name' : ', '.join(list(map(str, set(anndata_value.var.loc[:, 'gencode_v44_name'])))),
            'summed_genes' : True
        },
    )
    # Specify a new name for the collapsed gene axis
    ann_data_dict_sum[anndata_key].var_names = var_name

adata_summed_dups = ad.concat(
    list(ann_data_dict_sum.values()),
    axis= 1,
    join= 'outer',
    merge= 'unique'
)

# Remove all objects we no longer need
del duplicates_dict
del anndata_value
del anndata_key
del summed_expression
del var_name
del ann_data_dict_sum
del adata_dup

adata_summed_dups
```

We will now create a separate AnnData with only unique entries (to merge in following steps with our collapsed-entries AnnData)

```python
# Filter out duplicated var_names
adata_unique = adata_facs_10x_trimmed[:, ~adata_facs_10x_trimmed.var['gencode_v44_name'].isin(duplicate_var_names)].copy()
adata_unique.var['summed_genes'] = False
adata_unique, bool(adata_unique.var_names.duplicated().any())
```

We concat both AnnData objects to have a final one without duplicates and with full metadata

```python
adata_final = ad.concat(
    [adata_unique, adata_summed_dups],
    axis= 1,
    join= 'outer',
    merge= 'unique'
)
del adata_unique
del adata_summed_dups

adata_final
```

***NOTICE:*** the size of our original AnnData object compared to our final AnnData

There is only a 6-genes difference (the duplicates)

```python
adata_facs_10x_trimmed, adata_final
```

```python
# Write to data directory our annotated AnnData
adata_final.write_h5ad(
    './data/Travaglini2020/Travaglini2020_facs10x_AdultLung_annotated.h5ad',
    compression=hdf5plugin.FILTERS["zstd"],
    compression_opts=hdf5plugin.Zstd(clevel=5).filter_options
)

del adata_facs_10x_trimmed
del adata_final
```

## Normalization & Highly Variable Genes

```python
# Save raw data
adata_adult_lung.layers['counts'] = csr_matrix(adata_adult_lung.X)

# Normalize counts per cell.
# Normalize each cell by total counts over all genes, so that every cell has the same total count after normalization
sc.pp.normalize_total(adata_adult_lung, target_sum=1e4)

# Logarithmize the data matrix.
# Computes X = log (X=1), where X denotes the natural logarithm unless a different base is given.
sc.pp.log1p(adata_adult_lung)

adata_adult_lung.raw = adata_adult_lung.copy()
adata_adult_lung
```

## Scaling of the data

```python
# Scale data to unit variance and zero mean store results in layer
# Variables (genes) that do not display any variation (are constant across all observations) 
# are retained and set to 0 during this operation. In the future, they might be set to NaNs.

adata_adult_lung.layers["scaled"] = sc.pp.scale(adata_adult_lung, copy=True).X.copy()
adata_adult_lung
```
